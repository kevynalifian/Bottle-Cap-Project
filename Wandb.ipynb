```python
import wandb
from ultralytics import YOLO

# Setup wandb project
wandb.init(
    project="New_yolov8n_bottle_project",
    name="yolov8n-colab",
    config={
        "model": "yolov8n",
        "epochs": 100,
        "batch": 16,
        "imgsz": 640
    }
)

# Load model
model = YOLO("yolov8n.pt")

# Training
model.train(
    data="/content/bottle.yaml",
    epochs=100,
    batch=16,
    imgsz=640,
    project="New_yolov8n_bottle_project",
    name="caps_nano_run",
)

wandb.finish()
```
